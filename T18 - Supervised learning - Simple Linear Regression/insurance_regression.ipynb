```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
insurance_df = pd.read_csv("insurance.csv")

insurance_data = insurance_df
insurance_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = insurance_data.iloc[:,:1].values
y = insurance_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fit the model
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# Plot the data and model
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Playing around with the dataset
# Plot best fit line for NON SMOKERS
insurance_data_non_smoker = insurance_df[(insurance_df['smoker'] == "no")]
x_ns = insurance_data_non_smoker.iloc[:,:1].values
y_ns = insurance_data_non_smoker.iloc[:,-1].values

insurance_model_ns = LinearRegression()
insurance_model_ns.fit(x_ns,y_ns)
```

```python
y_ns_pred = insurance_model_ns.predict(x_ns)

plt.scatter(x,y,color = 'b')
plt.plot(x_ns,y_ns_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
